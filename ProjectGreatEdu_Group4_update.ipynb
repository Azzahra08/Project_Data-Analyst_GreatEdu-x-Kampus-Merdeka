```python
#Menghilangkan kolom yang tidak perlu
df2 = df1.drop(columns=['s.no', 'Ship Entry/Left','Type of ship present'])
df2
```

```python
df2.shape
```

```python
#Mereset index agar dimulai dari 0
df2.reset_index(drop=True, inplace=True)
df2
```

```python
#Mengubah nama kota Visakapatnam menjadi Visakhapatnam bersumber dari literatur
df2['City']=df2['City'].str.replace(r'Visakapatnam', 'Visakhapatnam', regex=True)
df2
```

```python
# Mengisi nilai NaN dalam kolom 'AQI' di df2
df3 = df2.copy()
df3['AQI'].fillna((df3['PM2.5 AQI'] + df3['NO2 AQI'] + df3['NH3 AQI'] + df3['CO AQI'] + df3['SO2 AQI'] + df3['O3 AQI'] + df3['VOC AQI']) / 7, inplace=True)
df3
```

```python
# Mengisi nilai NaN dalam kolom 'AQI_LVL' di df2
import numpy as np
df4 = df3.copy()
df4['AQI_LEVEL'] = np.select([
    (df4['AQI']<51),
    (df4['AQI'].between(50, 101)),
    (df4['AQI'].between(100, 151)),
    (df4['AQI'].between(150, 201)),
    (df4['AQI'].between(200, 301)),
    (df4['AQI'] > 300)
], ['GOOD', 'MODERATE', 'UNHEALTHY FOR SENSITIVE GROUPS', 'UNHEALTHY', 'VERY UNHEALTHY', 'HAZARDOUS'], default='undetermined')
```

```python
df5 = df4.drop(columns=['AQI_LVL'])
df5
```

```python
#Membuat grafik rata rata untuk data sebelum dideteksi outlier
import matplotlib.pyplot as plt
df_graf = df5.copy()

from matplotlib.dates import MonthLocator, DateFormatter
# Konversi kolom Date ke dalam tipe data datetime
df_graf['Date'] = pd.to_datetime(df_graf['Date'], format='%d-%m-%Y')

# Group by City dan Date, lalu hitung rata-rata AQI
df_grouped_graf = df_graf.groupby(['City', df_graf['Date'].dt.to_period("M")])['AQI'].mean().reset_index()

# Konversi Period ke datetime
df_grouped_graf['Date'] = df_grouped_graf['Date'].dt.to_timestamp()

# Pivot table untuk memisahkan data berdasarkan kota
df_pivot_graf = df_grouped_graf.pivot(index='Date', columns='City', values='AQI')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_pivot_graf.index, df_pivot_graf['Kolkata'], label='Kolkata', marker='o')
plt.plot(df_pivot_graf.index, df_pivot_graf['Visakhapatnam'], label='Visakhapatnam', marker='o')

# Label dan judul
plt.xlabel('Bulan')
plt.ylabel('Rata-rata AQI')
plt.title('Rata-rata AQI per Bulan - Visakhapatnam vs Kolkata', fontsize=16)
plt.legend()

# Atur format tanggal sumbu x
plt.gca().xaxis.set_major_locator(MonthLocator())
plt.gca().xaxis.set_major_formatter(DateFormatter('%b'))

# Tampilkan grafik
plt.show()
```

```python
# Melakukan encoding
df_a = df6.copy()
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

encoding_dict_aqilevl = {
    'GOOD': 0,
    'MODERATE': 1,
    'UNHEALTHY FOR SENSITIVE GROUPS': 2,
    'UNHEALTHY': 3,
    'VERY UNHEALTHY': 4,
    'HAZARDOUS': 5
}
encoding_dict_season = {
    'spring': 0,
    'summer': 1,
    'winter': 2,
}


df_a['AQI_LEVEL_Encod'] = df_a['AQI_LEVEL'].map(encoding_dict_aqilevl)
df_a['Season_Encod'] = df_a['Season'].map(encoding_dict_season)
df_a
```

```python
# Membuat korelasi menggunakan heatmap
import seaborn as sns
df_to_cor = df_a.drop(columns=['City','Season','Date','PM2.5 AQI CAT','NO2 AQI CAT','NH3 AQI CAT','CO AQI CAT',
                               'SO2 AQI CAT','O3 AQI CAT','VOC AQI CAT','AQI_LEVEL'])
korelasi = df_to_cor.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', linewidths=0.05)
plt.title('Correlation With Heatmap', fontsize=20)
plt.show()
```

```python
x = df_a[['PM2.5', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'VOC', 'AQI', 'Season_Encod']]
y = df_a['AQI_LEVEL_Encod']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_standar = scaler.fit_transform(x)
x_standar
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Teknik pemilihan data random sampling 
random_seed = 42
x_train, x_test, y_train, y_test = train_test_split(x_standar, y, test_size=0.2, random_state=random_seed)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
#Random Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)

train_score_rf = rf.score(x_train, y_train)
test_score_rf = rf.score(x_test, y_test)

print(f"RF on the training dataset : {train_score_rf:.2%}")
print(f"RF on the test dataset : {test_score_rf:.2%}")
```

```python
#SVM
svm = SVC()
svm.fit(x_train, y_train)
svm_pred = svm.predict(x_test)

train_score_svm = svm.score(x_train, y_train)
test_score_svm = svm.score(x_test, y_test)

print(f"SVM on the training dataset: {train_score_svm:.2%}")
print(f"SVM on the test dataset: {test_score_svm:.2%}")
```

```python
#XGBoost
xgb_mod = xgb.XGBClassifier()
xgb_mod.fit(x_train, y_train)
xgb_pred = xgb_mod.predict(x_test)

train_score_xgb = xgb_mod.score(x_train, y_train)
test_score_xgb = xgb_mod.score(x_test, y_test)

print(f"XGB on the training dataset : {train_score_xgb:.2%}")
print(f"XGB on the test dataset : {test_score_xgb:.2%}")
```

Overfitting adalah kondisi dimana nilai akurasi pada data training tinggi dengan nilai akurasi data testing rendah. Kondisi best fitting merupakan kondisi yang paling ideal antara nilai akurasi pada data testing dan data training sama-sama tinggi. Maka dari itu, untuk membangun model machine learning pilihlah kondisi best fitting. Meskipun memang sama-sama tinggi tapi nilai bias dan variansinya rendah.
